```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import *
from pyspark.sql.window import Window
```

```python
df = spark.table('workspace.bronze_schema.cust_az_delta')
df.display()
```

```python
df = df.fillna('Not Available', subset=['GEN'])
df.display()
```

### TRIM FOR SPACES

```python
df = df.withColumn('GEN',trim(col('GEN')))
df.display()
```

### REPLACE/ NORMALIZE COLUMN NAME


```python
df = df.withColumn('GEN', when(col('GEN') == 'F', 'Female').otherwise(col('GEN')))\
    .withColumn('GEN', when(col('GEN') == 'M', 'Male').otherwise(col('GEN')))\
        .withColumn('GEN', when(col('GEN') == '', 'Not Available').otherwise(col('GEN')))

df.display()
```

### WRITE TO SILVER SCHEMA

```python
df.write.mode('overwrite').saveAsTable('workspace.silver_schema.cust_az_silver')
```
